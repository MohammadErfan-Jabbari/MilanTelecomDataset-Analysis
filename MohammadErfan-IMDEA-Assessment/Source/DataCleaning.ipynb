In this file we will read the original txt dataset files and then we will only keep the necessary columsn, save them in new files. Why we don't concat them into a big signle file, it's in the task I we will need to calculate some features only for sime specific days.

```python
# necassary imports
import pandas as pd
import os
```

```python
def cleanData(fileAddress, selectedColumns, destinationAddress):
    df = pd.read_csv(fileAddress + '.txt', sep='\t', names=['square_id', 'time_interval', 'country_code', 'sms_in', 'sms_out', 'call_in', 'call_out', 'internet_traffic'])
    df.fillna(0, inplace=True)
    # Filter selected columns
    selected_data = df[selectedColumns]
    # Remove multiple items for same timestamp
    df = selected_data.groupby(['square_id', 'time_interval']).sum().reset_index()
    # convert to date_time
    df['time_interval'] = pd.to_datetime(df['time_interval'], unit="ms")
    df.set_index('time_interval', inplace=True)
    # Convert to hourly data
    new_df = df.groupby('square_id').resample('H')['internet_traffic'].agg('sum').loc[lambda x: x>0]

    new_df = new_df.reset_index()

    new_df.to_csv(destinationAddress + '.csv', index = False)
    

    # df['time_interval'] = pd.to_datetime(df['time_interval'], unit="ms")
    # df.set_index('time_interval', inplace=True)
    # # resample the data into half-hourly intervals
    # df_half_hourly = df.resample('30T').sum()

    # df_half_hourly.to_csv(destinationAddress + '.csv', index = False)
```

```python
# cleanData('sms-call-internet-mi-2013-11-01', 'txt', ['square_id', 'time_interval', 'internet_traffic'])


# Define the folder path
txt_dataset_folder = './txt_dataset/'
cleaned_dataset_folder = './cleaned_dataset_30/'
selected_columns = ['square_id', 'time_interval', 'internet_traffic']

# Get the list of files in the folder
file_list = os.listdir(txt_dataset_folder)

# Print the list of files
for file in file_list:
    file_name = os.path.splitext(file)[0]
    cleanData(txt_dataset_folder + file_name, selected_columns, cleaned_dataset_folder + 'cleaned-' + file_name)
    print(cleaned_dataset_folder + 'cleaned-' + file_name + 'finished \n')
```
